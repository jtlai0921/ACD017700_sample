## 3.2 試用CartPole

```python
# 宣告使用的套件
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import gym
```

```python
# 宣告動畫的繪圖函數
# 参考URL http://nbviewer.jupyter.org/github/patrickmineault/xcorr-notebooks/blob/master/Render%20OpenAI%20gym%20as%20GIF.ipynb
from JSAnimation.IPython_display import display_animation
from matplotlib import animation
from IPython.display import display


def display_frames_as_gif(frames):
    """
    Displays a list of frames as a gif, with controls
    """
    plt.figure(figsize=(frames[0].shape[1]/72.0, frames[0].shape[0]/72.0),
               dpi=72)
    patch = plt.imshow(frames[0])
    plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = animation.FuncAnimation(plt.gcf(), animate, frames=len(frames),
                                   interval=50)

    anim.save('movie_cartpole.mp4')  # 註記：儲存為動畫的部分
    display(display_animation(anim, default_mode='loop'))
```

```python
# 讓CartPole隨機移動

frames = []
env = gym.make('CartPole-v0')
observation = env.reset()  # 一開始必須先重置環境

for step in range(0, 200):
    frames.append(env.render(mode='rgb_array'))  # 於frames追加每個時間點的影像
    action = np.random.choice(2)  # 隨機傳回0(推車往左), 1(推車往右)
    observation, reward, done, info = env.step(action)  # 執行action

# 注意：執行後，會開啟ipykernel_launcher.p這個視窗，請先放著，不用多加理會
```

```python
# 儲存與繪製動畫
display_frames_as_gif(frames)
```
